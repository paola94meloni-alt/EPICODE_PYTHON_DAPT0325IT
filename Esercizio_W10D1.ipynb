```python
#Scrivere un programma che 
#• in input acquisisce una lista di numeri e un numero K 
#• in output, dovrà restituire la media di tutti i numeri nella lista maggiori o uguali a K 
#• se non ce ne dovesse essere nessuno, dovrà stampare a schermo un messaggio adeguato.
lista = [2, 1000, 350, 7, 28, 49, 678, 0, 89, 412, 5421, 862, 7648]
K = 80
lista_array = np.array(lista)
filtro = lista_array >= K
lista_array_filtrata = lista_array [filtro]
print (lista_array_filtrata) #numeri >= K
lista_array_filtrata.mean() #media dei numeri >= K
if len(lista_array_filtrata) == 0:
    print ("non ci sono numeri maggiori o uguali a 80")
else:
    print ("la media dei numeri >= a K è", lista_array_filtrata.mean())
```

```python
#Abbiamo una lista di stringhe di prezzi in dollari, che erroneamente sono stati scritti con il simbolo dell'euro: 
prezzi = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"] 
#cambiare il simbolo dell'euro (€) in quello del dollaro ($) per ogni stringa nella lista; 
#il risultato sarà memorizzato in un'altra lista.
prezzi_agg = []
for prezzo in prezzi:
    prezzo_aggiustato = prezzo.replace("€" , "$")
    print(prezzo_aggiustato)
    prezzi_agg.append(prezzo_aggiustato)
print(prezzi_agg)
```

```python
#Abbiamo una lista con i guadagni degli ultimi 12 mesi (supponiamo da Gennaio a Dicembre): 
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50] 
#dobbiamo confrontare, stampando tutto a video, il guadagno di ogni mese con la media dei guadagni precedenti, 
#e specificare nell'output se il guadagno attuale è maggiore o minore della media dei precedenti.
import numpy as np
for i, guadagno in enumerate(guadagni):
    if i == 0:
        print("Mese 1:", guadagno, "€ — non ci sono mesi precedenti\n")
        continue
    media = np.mean(guadagni[0:i])
    diff = guadagno - media
    print("Guadagno mese ", i , ": ", guadagno, " €")
    print("Media mesi precedenti: ", media, " €")
    if diff > 0:
        print("il guadagno attuale è maggiore della media dei guadagni dei mesi precedenti\n")
    else:
        print("il guadagno attuale è minore della media dei guadagni dei mesi precedenti\n")


#Esempio di un possibile output: 
#Mese 1: 100 € Mese 2: 90 € (media prec: 100 € - il guadagno attuale è minore) 
#Mese 3: 70 € (media prec: 95 € - il guadagno attuale è minore)
```
