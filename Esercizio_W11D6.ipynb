```python
#Dal database AdventureWorks estraiamo la tabella dimproduct 
#• Sulla colonna DealerPrice, utilizzando il metodo .round(), arrotondiamo i valori alle due cifre decimali, 
#e poi al valore intero più vicino 
#• Utilizzando il metodo .clip(), facciamo in modo che i valori siano compresi tra un minimo di 0 e un massimo di 1000
import os
import dotenv
import sqlalchemy
import pandas as pd
import numpy as np

dotenv.load_dotenv(override=True, dotenv_path="credenziali.env")

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(username, password, host, dbname)
db_engine = sqlalchemy.create_engine(connection_string)
query = "SELECT * FROM dimproduct"
dimproduct_df = pd.read_sql(query, db_engine)
display (dimproduct_df)
```

```python
#• Sulla colonna DealerPrice, utilizzando il metodo .round(), arrotondiamo i valori alle due cifre decimali, 
#e poi al valore intero più vicino
dimproduct_df["DealerPrice"]= dimproduct_df["DealerPrice"].round(2)
print(dimproduct_df["DealerPrice"])
```

```python
#e poi al valore intero più vicino
dimproduct_df["DealerPrice"] = dimproduct_df["DealerPrice"].round(0)
print(dimproduct_df["DealerPrice"])
```

```python
#• Utilizzando il metodo .clip(), facciamo in modo che i valori siano compresi tra un minimo di 0 e un massimo di 1000
dimproduct_df_clip = dimproduct_df["DealerPrice"].clip(lower=0, upper=1000)
print(dimproduct_df_clip)
```

```python
#Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate, 
#con il seguente codice: 
years = 5 
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), 
                         "Anno": np.repeat(list(range(years)), 12), 
                         "Valore": np.random.randint(800, 5000, 12*years)}) 
#• Calcola la somma cumulativa dei guadagni utilizzando il metodo .cumsum() 
#• Come sopra, ma raggruppato per ogni anno usando prima un .groupby()
display(guadagni)
cumulativo_totale= guadagni["Valore"].cumsum()
cumulativo_per_anno= guadagni.groupby("Anno")["Valore"].sum()
print("cumulativo totale", cumulativo_totale)
print("cumulativo_per_anno", cumulativo_per_anno)
```

```python
#Dai beginner_datasets carichiamo in un DataFrame il file facebook.csv, che contiene dei post con data di pubblicazione, 
#tipo (foto, video, …) e numero di reactions raccolte: 
#• Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp 
#• Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, 
#ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, 
#il giorno dell'anno, eccetera.
face = pd.read_csv("facebook.csv")
print(face.shape)
display(face.sample(10))
print(face.columns)
```

```python
#• Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp 
#• Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, 
#ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, 
#il giorno dell'anno, eccetera.
face["status_published"] = pd.to_datetime(face["status_published"])
face["Year"] = face.loc [:,"status_published"].dt.year
face["Month"] = face.loc [:,"status_published"].dt.month
face["Day"] = face.loc [:,"status_published"].dt.day
face["DayOfWeek"] = face.loc [:,"status_published"].dt.dayofweek
face["DayOfYear"] = face.loc [:,"status_published"].dt.dayofyear
print (face["Year"] , face["Month"] , face["Day"] , face["DayOfWeek"] , face["DayOfYear"])
```

```python
#• Estraiamo solo i post relativi al 2012 
#• Estraiamo solo i post relativi a maggio 2018 
#• Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana 
#• Troviamo il primo e ultimo post pubblicati in ogni anno 
#• Quanti tipi di post ci sono? E quanti per ogni tipo?

#• Estraiamo solo i post relativi al 2012 
filtro = face["Year"] == 2012
post_2012 = face.loc[filtro]
print (post_2012)


#• Estraiamo solo i post relativi a maggio 2018 
filtro2 = face["Year"] == 2018
filtro3 = face["Month"] == 5
post_may_2018 = face.loc[filtro2 & filtro3]
print (post_may_2018)


#• Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana 
filtro4 = face["DayOfWeek"].isin([0, 1, 2, 3, 4])
filtro5 = face["DayOfWeek"].isin([5, 6])

post_settimana = face[filtro4].shape[0]
post_weekend = face[filtro5].shape[0]
print("Post della settimana:", post_settimana)
print("Post del weekend:", post_weekend)


#• Troviamo il primo e ultimo post pubblicati in ogni anno
for year in face["Year"].unique():
    df_year = face[face["Year"] == year]  # Filtra per l'anno
    primo_post = df_year.head(1)          # Primo post dell'anno
    ultimo_post = df_year.tail(1)         # Ultimo post dell'anno
    print(f"Anno: {year} \nPrimo post: {primo_post} \nUltimo post: {ultimo_post}")


#• Quanti tipi di post ci sono? E quanti per ogni tipo?
tipologia_post = face["status_type"].unique()
print ("Le tipologie di post sono " , tipologia_post)
conteggio_post = face["status_type"].value_counts()
print("Conteggio post per ogni tipologia \n" , conteggio_post)
```

```python
#Dai beginner_dataset carichiamo in un DataFrame il file automobile.csv: 
#• Ci sono valori nulli? Dove? Quanti? 
#• Quali righe hanno un valore nullo nella colonna num-of-doors? 
#• Esaminando i dati nel dataset, cerchiamo una logica per sostituire i valori nulli nella colonna num-of-doors

auto = pd.read_csv("automobile.csv")
print(auto.shape)
display(auto.sample(10))
auto.info()

print(auto.isnull().sum())
#ci sono 37 valori nulli nella colonna "normalized-losses" e 2 valori nulli nella colonna "num-of-doors"

#• Quali righe hanno un valore nullo nella colonna num-of-doors? 
filtro= auto["num-of-doors"].isna()
righe_nulle = auto.loc[filtro]
display(righe_nulle)

#• Esaminando i dati nel dataset, cerchiamo una logica per sostituire i valori nulli nella colonna num-of-doors
print(auto.groupby("body-style")["num-of-doors"].value_counts())
#le sedan hanno per la maggiore 4 porte, perciò è probabile che i dati mancanti siano proprio "four"!
auto["num-of-doors"] = auto["num-of-doors"].fillna("four")

```

```python
lista_file = os.listdir("datasets")
for nome_file in lista_file:
    if ("._" not in nome_file) and (".csv" in nome_file):
        df = pd.read_csv ("datasets/" + nome_file)
        numero_valori_nulli = df.isna().sum().sum()
        if numero_valori_nulli > 0:
            print (nome_file, "\t", numero_valori_nulli)
```
