```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(override=True, dotenv_path="credenziali.env")

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(username, password, host, dbname)
db_engine = sqlalchemy.create_engine(connection_string)
query = "SELECT * FROM dimproduct"
df = pd.read_sql(query, db_engine)
print (df)
```

```python
# 1 In media quanto pesano i prodotti 
#(per questo esercizio, ignoriamo l'unità di misura, ma usiamo solo i valori della colonna relativa)? 
#• Quanto pesa il più leggero? • Quanto pesa il più pesante? 
#• Quanti prodotti pesano più di 100? 
#• Quanto costano in media i prodotti (colonna DealerPrice)? • Quali sono i quartili dei prezzi?
# 1 In media quanto pesano i prodotti 
print ("- media dei pesi" , df["Weight"].mean().round(2)) #round per arrotindare il numero tra parentesi metto quante cifre decimali voglio
# 2 quanto pesa il più leggero?
print("- il prodotto più leggero pesa" , df["Weight"].min())
# 3 quanto pesa il più pesante?
print("- il prodotto più pesante pesa" , df["Weight"].max())
# 4 quanti prodotti pesano più di 100?
filtro = df["Weight"]>100
print("- i prodotti che pesano più di 100 sono" , df.loc[filtro].shape[0]) #sono 28 righe, quindi 28 oggetti
# 5 quanto costano in media i prodotti?
print ("- il costo medio dei prodotti è" , df["DealerPrice"].mean().round(2))
# 6 quali sono i quartili dei prezzi?
q1 = df["ListPrice"].quantile(0.25)
q2 = df["ListPrice"].quantile(0.50)   # mediana
q3 = df["ListPrice"].quantile(0.75)
print("- Primo quartile (Q1):", q1)
print("- Secondo quartile (Q2):", q2)
print("- Terzo quartile (Q3):", q3)
#oppure si può fare con describe che descrive la colonna statisticamente
df["ListPrice"].describe()
```

```python
#Qual è il prezzo medio per i prodotti di colore blu?
filtro = df["Color"] == "Blue"
prezzo_medio_prod_blu = df[filtro]["ListPrice"].mean()
print("- il prezzo medio dei prodotti di colore Blu è" , prezzo_medio_prod_blu)
#oppure posso farlo con loc:
prezzo_medio_prod_blu = df.loc[df["Color"] == "Blue", "ListPrice"].mean()
print("- il prezzo medio dei prodotti di colore Blu è" , prezzo_medio_prod_blu)
#• Qual è il prezzo medio per i prodotti di colore rosso o nero?
filtro = (df["Color"] == "Red") | (df["Color"] == "Black")
prezzo_medio_rosso_o_nero = df[filtro]["ListPrice"].mean()
print("- il prezzo medio dei prodotti di colore Rosso o Nero è" , prezzo_medio_rosso_o_nero)
#oppure posso farlo con loc
rosso = (df["Color"] == "Red")
nero = (df["Color"] == "Black")
prezzo_medio_rosso_o_nero = df.loc[(rosso | nero) , "ListPrice"].mean()
print("- il prezzo medio dei prodotti di colore Rosso o Nero è" , prezzo_medio_rosso_o_nero)
#• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?
filtro = (df["Weight"] > 10) & (df["Size"] == "42")
prezzo_max_tg_42_peso_maggiore_10 = df[filtro]["ListPrice"].max()
print("- il prezzo massimo di prodotti di tg 42 e peso > di 10 Kg è", prezzo_max_tg_42_peso_maggiore_10)
#oppure posso farlo con loc:
peso = (df["Weight"] > 10)
taglia = (df["Size"] == "42")
prezzo_max_tg_42_peso_maggiore_10 = df.loc [(peso & taglia) , "ListPrice"].max()
print("- il prezzo massimo di prodotti di tg 42 e peso > di 10 Kg è", prezzo_max_tg_42_peso_maggiore_10)
#• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg 
#e colore argento? 
filtro = (df["Weight"] > 10) & (df["Size"] == "42") & (df["Color"] == "Silver")
risultato = df.loc[filtro, ["EnglishProductName", "StandardCost"]]
print(risultato)
#• Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: quali sono le differenze? 
#C'è un pattern? C'è qualche elemento che non lo segue?
ultimi_20 = df.tail(20)
print(ultimi_20[["EnglishProductName", "StandardCost","DealerPrice"]])
```

```python
#Il dataset diabetes.csv raccoglie persone con diabete o meno, e il valore di diverse variabili fisiologiche dei pazienti. 
#• Osserviamone le dimensioni e un'anteprima di cinque righe; 
#• Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne; 
#• Stampiamo dei descrittori statistici del dataset; 
#• Selezioniamo i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50; 
#• Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
df = pd.read_csv("diabetes.csv")
df
#• Osserviamone le dimensioni e un'anteprima di cinque righe; 
print(df.shape)
display(df.sample(5))
#• Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne; 
print(df.info())
print (df.columns)
#• Stampiamo dei descrittori statistici del dataset;
df.describe()
```

```python
#• Selezioniamo i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50;
eta_20_29 = df[(df["Age (years)"] >=20) & (df["Age (years)"] <=29)]
eta_30_39 = df[(df["Age (years)"] >=30) & (df["Age (years)"] <=39)]
eta_40_50 = df[(df["Age (years)"] >=40) & (df["Age (years)"] <=50)]
display (eta_20_29)
display (eta_30_39)
display (eta_40_50)
```

```python
#• Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
filtro1 = (df["Age (years)"] >=20) & (df["Age (years)"] <=29)
filtro2 = (df["Age (years)"] >=30) & (df["Age (years)"] <=39)
filtro3 = (df["Age (years)"] >=40) & (df["Age (years)"] <=50)

press_diast_20_29 = df[filtro1]["Diastolic blood pressure (mm Hg)"].mean().round(2)
press_diast_30_39 = df[filtro2]["Diastolic blood pressure (mm Hg)"].mean().round(2)
press_diast_40_50 = df[filtro3]["Diastolic blood pressure (mm Hg)"].mean().round(2)
print ("media_press_diast_20_29 = ", press_diast_20_29)
print ("media_press_diast_30_39 = ", press_diast_30_39)
print ("media_press_diast_40_50 = ", press_diast_40_50)
```

```python
#Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini delle persone, 
#e della zona in cui vivono, rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute.
# Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
# Quali sono le medie di charges rispetto a region? 
# Ci sono differenze significative? 
# E rispetto a smoker? e a sex?
# Quali sono i descrittori statistici di bmi? 
# Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
# Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna;
df = pd.read_csv("insurance.csv")
print(df.shape)
display(df.sample(10))
print(df.columns)
# Quali sono le medie di charges rispetto a region? 
filtro1 = (df["region"] == "northwest") & (df["smoker"] == "yes")
filtro2 = df["region"] == "southwest"
filtro3 = df["region"] == "northeast"
filtro4 = df["region"] == "southeast"
media_charges_northwest = df.loc[filtro1, "charges"].mean().round()
media_charges_southwest = df.loc[filtro2, "charges"].mean().round()
media_charges_northeast = df.loc[filtro3, "charges"].mean().round()
media_charges_southeast = df.loc[filtro4, "charges"].mean().round()
print("Media charges Northwest:", media_charges_northwest)
print("Media charges Southwest:", media_charges_southwest)
print("Media charges Northeast:", media_charges_northeast)
print("Media charges Southeast:", media_charges_southeast)
# Ci sono differenze significative?
#In realtà non ci sono differenze così significative, le medie per le regioni a east sono un po' più alte ma non di troppo.
# E rispetto a smoker? e a sex?
filtro1 = df["region"] == "northwest"
filtro2 = df["region"] == "southwest"
filtro3 = df["region"] == "northeast"
filtro4 = df["region"] == "southeast"
media_smoker_northwest = (df.loc[filtro1, "smoker"] == "yes").mean().round(2)
media_smoker_southwest = (df.loc[filtro2, "smoker"] == "yes").mean().round(2)
media_smoker_northeast = (df.loc[filtro3, "smoker"] == "yes").mean().round(2)
media_smoker_southeast = (df.loc[filtro4, "smoker"] == "yes").mean().round(2)

print("Percentuale smoker Northwest:", media_smoker_northwest)
print("Percentuale smoker Southwest:", media_smoker_southwest)
print("Percentuale smoker Northeast:", media_smoker_northeast)
print("Percentuale smoker Southeast:", media_smoker_southeast)
```

```python
#Il dataset pokemon.csv contiene un database di Pokémon, con dati quali nome, tipi di attacco, valori di attacco/difesa/eccetera, 
#e se sono o meno leggendari. 
#• Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna; 
#• È verosimile che la prima colonna dovrebbe essere un indice? #SI E' UN INDICE
#• Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? #NO, NON COMBACIANO
#• Se no, settare la prima colonna come indice.
#• Quali sono i Pokémon leggendari?
#• E quali sono i leggendari di tipo 1 Grass? 
#• E leggendari di tipo 1 Ice o Fire?
#• Trasformiamo la colonna Name nell'indice; 
#• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
import pandas as pd
df = pd.read_csv("pokemon.csv")
```

```python
#• Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna;
print(df.shape)
display(df.sample(10))
display(df.head(15))
print(df.columns)
```

```python
#• È verosimile che la prima colonna dovrebbe essere un indice? #SI E' UN INDICE
#• Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? #NO, NON COMBACIANO
#• Se no, settare la prima colonna come indice.
df = df.set_index("#")
display (df)
```

```python
#• Quali sono i Pokémon leggendari?
filtro = df["Legendary"] == True
display(df.loc[filtro])
```

```python
#• E quali sono i leggendari di tipo 1 Grass? 
filtro1 = df["Legendary"] == True
filtro2 = df["Type 1"] == "Grass"
display(df.loc[filtro1 & filtro2])
```

```python
#• E leggendari di tipo 1 Ice o Fire?
filtro1 = df["Legendary"] == True
filtro3 = df["Type 1"] == "Ice"
filtro4 = df["Type 1"] == "Fire"
display(df.loc[filtro1 & filtro3 | filtro4])
```

```python
#• Trasformiamo la colonna Name nell'indice; 
df_name= df.set_index("Name")
display(df_name)
```

```python
#• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
filtro1 = df["Generation"] == 1
filtro2 = df["Attack"] > 50
filtro3 = df["HP"] < 60
display(df.loc[filtro1 & filtro2 & filtro3])
```
