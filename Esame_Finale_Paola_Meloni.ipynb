```python
import pandas as pd
import numpy as np
import seaborn as sns
url =  "https://catalog.ourworldindata.org/garden/covid/latest/compact/compact.csv"
df = pd.read_csv(url)
df
display(df.head(50))
```

```python
# 1 Si richiede di verificare le dimensioni del dataset e i relativi metadati
print("Numero Righe e numero Colonne:", df.shape)
df.info() #metadati
df.describe() #per completare l'EDA
```

```python
# 2 Si chiede di trovare, per ogni continente: 
# a. il numero di casi fin dall'inizio della pandemia 
# b. la percentuale rispetto al totale mondiale del numero di casi

#per svolgere questa richiesta devo usre le colonne "continent" e "new_cases" perchÃ¨ se uso la colonna total cases (che Ã¨ cumulativa)
#continuo a sommare sempre gli stessi contagi.
```

```python
#controllo se ci sono nulli nelle due colonne
filtro1 = df["continent"].isna()
righe_nulle_cont = df.loc[filtro1]
display(righe_nulle_cont.head(50))
display(righe_nulle_cont)
#ci sono 40960 nulli in "continent"

filtro2 = df["new_cases"].isna()
righe_nulle_new_cases = df.loc[filtro2]
display(righe_nulle_new_cases)
#ci sono 16752 nulli in "new_cases" 
```

```python
display(righe_nulle_cont[["continent" , "country"]].sample(50))
```

```python
#in questo caso mi sono accorta che posso inserire in continent il valore della colonna country
#ma che in alcuni casi in country ci sono dei dati "strani" per esempio "World".
#capisco quali valori di continent mi servono e poi uguaglio le colonne di country e continent
continenti = df["continent"].unique()
print(continenti)
```

```python
filtro3 = df["continent"].isna() #filtro per i continent che sono Nan
continenti = ['Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America']
filtro4 = df["country"].isin(continenti) #filtro per le righe in cui country Ã¨ uno dei continenti "reali"

df.loc[filtro3 & filtro4, "continent"] = df.loc[filtro3 & filtro4, "country"] 
#se filtro1 e filtro2 sono veri entrambi assegno il valore di country a continent

#verifico che ora i valori di continent che erano Nan siano stati sostituiti con il valore presente in country SOLO 
#dove in country c'era il nome di un continente.
filtro5 = df["continent"].isna()
righe_nulle_cont = df.loc[filtro5]

display(righe_nulle_cont[["continent" , "country"]].sample(50))
```

```python
#cancello i Nan da continent sovrascrivendo la variabile iniziale
df = df[df["continent"].notna()]
display(df)
```

```python
# a. il numero di casi fin dall'inizio della pandemia
#raggruppo per continente e aggrego i dati facendo la somma di new_cases
numero_casi_tot_per_continente = df.groupby("continent")["new_cases"].sum() 
print (numero_casi_tot_per_continente)
```

```python
# b. la percentuale rispetto al totale mondiale del numero di casi
#calcolo i casi totali mondiali sommando tutti i valori nella colonna new_cases
tot_casi_mondiali = df["new_cases"].sum()

#calcolo la percentuale raggruppando per continente (vedi cella precedente)
percentuali_per_continente = (numero_casi_tot_per_continente/tot_casi_mondiali)*100
print(percentuali_per_continente.round(2))
```

```python
# 3 Selezionare i dati relativi all'Italia nel 2022 e, poichÃ© i nuovi casi vengono registrati settimanalmente, 
#filtrare via i giorni che non hanno misurazioni; quindi mostrare con dei grafici adeguati: 
# a l'evoluzione del casi totali dall'inizio alla fine dell'anno 
# b. il numero di nuovi casi rispetto alla data

#trasformo la colonna date in formato data perchÃ¨ Ã¨ un object quindi intuisco che sia una stringa
df["date"] = pd.to_datetime(df["date"])

#estraggo l'anno dalla colonna date
df["Year"] = df.loc [:,"date"].dt.year
```

```python
# a l'evoluzione dei casi totali dall'inizio alla fine dell'anno 

#ripulisco il DF escludendo i giorni in cui non ci sono misurazioni di casi
df = df[df["total_cases"].notna()]

#filtro i dati per ITALIA e 2022

filtro6 = df["country"] == "Italy"
filtro7 = df["Year"] == 2022 
casi_italia_2022 = df.loc[filtro6 & filtro7]

#utilizzo un line plot perchÃ¨ Ã¨ quello che mostra meglio le variazioni nel tempo di una variabile
casi_italia_2022.plot(x="date", y="total_cases")
#il grafico mostra l'andamento dei casi TOTALI annuali del 2022, Ã¨ ovvio che alla fine dell'anno il numero dei casi sia molto
#piÃ¹ elevato che all'inizio perchÃ¨ la variabile Total_case Ã¨ cumulativa. Non si valuta un delta di valori, ma si valuta la costante
#somma di essi.
```
