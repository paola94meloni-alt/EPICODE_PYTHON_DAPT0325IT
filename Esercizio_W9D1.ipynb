```python
#Abbiamo la seguente stringa: my_string = "I am studying Python" • Trasformarla in modo che tutti i caratteri siano maiuscoli (uppercase) •
#Trasformarla in modo che tutti i caratteri siano minuscoli (lowercase) • Sostituire la sottostringa "Python" con la stringa "a lot"
#• Usare il metodo .strip(); cambia qualcosa? Perché? • Usare il metodo .split() per trovare tutte le parole (token)
my_string = "I am studying Python"
my_string_upper = my_string.upper()
print ("Uppercase: ", my_string_upper)
my_string_lower = my_string.lower()
print ("Lowercase: ", my_string_lower)
my_string_replace = my_string.replace ("Python" , "a lot")
print ("Replece: " , my_string_replace)
my_string_strip = my_string.strip()
print ("Strip: " , my_string_strip)
my_string_split = my_string.split()
print ("Split: " , my_string_split)
#usando strip non cambia nulla perchè in questo caso non ho spazi vuoti all'inizio o alla fine della stringa.
```
