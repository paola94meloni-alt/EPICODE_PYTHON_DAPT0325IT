```python
import pandas as pd
import numpy as np
df = pd.read_csv("Mappa-dei-pub-circoli-locali-in-Italia.csv",encoding="latin1", sep=";")
```

```python
#• quanti dati ci sono in totale? 
#• quali sono i metadati? • stampiamo il primo elemento 
#• stampiamo l'ultimo elemento • riusciamo a stampare un elemento a caso? 
print(df.shape)
print(df.info())
display(df.head(1))
display(df.tail(1))
display(df.sample(1))
```

```python
#• quali sono gli anni di inserimento presenti?
df["Anno inserimento"].unique()
```

```python
#• quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46? 
filtro1 = (df["Longitudine"] >= 9) & (df["Longitudine"] <= 10)
filtro2 = (df["Latitudine"] >= 45) & (df["Latitudine"] <= 46)
attivita_quad = df.loc[filtro1 & filtro2]
display(attivita_quad.shape[0])
```

```python
#• quante attività ci sono nella provincia di Vicenza? 
filtro3 = df["Provincia"] == "VICENZA"
attivita_vicenza = df.loc[filtro3]
display(attivita_vicenza.shape[0])
```

```python
#• quante enoteche ci sono, e come si chiamano?
enoteche = []
for nome in df["Nome"]:
    if pd.notna(nome) and "enoteca" in nome.lower():
        enoteche.append(nome)
print(len(enoteche))
display(enoteche)
```

```python
#• quante attività ci sono in Lazio e Abruzzo assieme?
filtro1 = df["Regione"] == "Lazio" 
filtro2 = df["Regione"] == "Abruzzo"
attivita_Lazio_Abruzzo = df.loc[filtro1 | filtro2]
print(attivita_Lazio_Abruzzo.shape[0])
```

```python
#Il dataset insurance.csv dei beginner_datasets contiene dati su caratteristiche e abitudini delle persone 
#rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute.
#• Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
#• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
#• E rispetto a smoker? E a sex? 
#• Quali sono i descrittori statistici di bmi? 
# Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi? 
#Nota: è lo stesso esercizio della volta scorsa, ma stavolta da eseguire con il metodo .groupby()
ins = pd.read_csv("insurance.csv")
print(ins.shape)
display(ins.sample(10))
print(ins.columns)
```

```python
#• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
medie_charges_per_regione = ins.groupby("region")["charges"].mean().round(2)
print(medie_charges_per_regione)
```

```python
#Carichiamo in un DataFrame il dataset iris.csv dei beginner_datasets e calcoliamo: 
#• La media della lunghezza dei petali di tutto il dataset 
#• La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby() 
#• Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()
iris = pd.read_csv("iris.csv")
print(iris.shape)
display(iris.sample(10))
print(iris.columns)
```

```python
#• La media della lunghezza dei petali di tutto il dataset 
print(iris["petal_length"].mean().round(2))
```

```python
#• La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby()
mean_lung_pet = iris.groupby("species")["petal_length"].mean().round(2)
print(mean_lung_pet)
```

```python
#• Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()
stat_sep = iris.groupby("species")["sepal_width"].agg(["mean","min","max"])
print(stat_sep)
```

```python
#Nei beginner_datasets, il dataset wine.csv contiene delle analisi organolettiche su diverse qualità di vini: 
#• Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale? 
#• C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi? 
#• Rifacendo le analisi dei due punti precedenti ma per il pH, cambia qualcosa? 
#• E per i solfati?
wine = pd.read_csv("wine.csv")
print(wine.shape)
display(wine.sample(15))
print(wine.columns)
```

```python
#• Qual è la media di concentrazione alcolica per ogni qualità?
#Ci sono differenze? E rispetto alla media totale?
conc_med_alc = wine.groupby("quality")["alcohol"].mean().round(2)
print (conc_med_alc)

#dv std? per vedere le differenze tra le varie qualità

conc_med_tot = wine["alcohol"].mean().round(2)
print (conc_med_tot)

diff_rispetto_tot = conc_med_alc - conc_med_tot
print (diff_rispetto_tot)
```

```python
#• Rifacendo le analisi dei due punti precedenti ma per il pH, cambia qualcosa? 
pH_med = wine.groupby("quality")["pH"].mean().round(2)
print (pH_med)

pH_med_tot = wine["pH"].mean().round(2)
print (pH_med_tot)

diff_rispetto_tot = pH_med - pH_med_tot
print (diff_rispetto_tot)
```

```python
#• E per i solfati?
solfati_med = wine.groupby("quality")["sulphates"].mean().round(2)
print (solfati_med)

solfati_med_tot = wine["sulphates"].mean().round(2)
print (solfati_med_tot)

diff_rispetto_tot = solfati_med - solfati_med_tot
print (diff_rispetto_tot)
```

```python
#Nei beginner_datasets, il file boston.csv contiene il Boston Housing Dataset, 
#che deriva dalle informazioni raccolte dal Census Service degli Stati Uniti sulle abitazioni 
#nell'area di Boston. 
#• La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 
#• Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato? 
#• Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? 
#Appare esserci una qualche correlazione? Come si può spiegare il risultato? 
#• Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? 
#E delle tasse?
bost = pd.read_csv("boston.csv")
print(bost.shape)
display(bost.sample(15))
print(bost.columns)
```

```python
#• La media del prezzo delle case cambia a seconda della distanza dal fiume Charles?
med_house_river = bost.groupby("chas")["medv"].mean().round(2)
print (med_house_river)
#cambia in media di 6.000 dollari
```

```python
#• Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato?
#df["col1"].corr(df["col2"])
correlation = bost["crim"].corr(bost["medv"])
print (correlation)
#i due sono inversamente correlati perchè il coefficiente di correlazione è negativo, ciò significa che al crescere del tasso
#di criminalità diminuisce il prezzo delle abitazioni e viceversa.
```

```python
#• Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? 
#Appare esserci una qualche correlazione? Come si può spiegare il risultato?
med_rooms = bost.groupby("ptratio")["rm"].mean().round(2)
print (med_rooms)
correlation1 = bost["ptratio"].corr(bost["rm"])
print (correlation1) #coeff. correlazione negativo = sono inversamente proporzionali, se aumenta il numero delle stanze,
#aumenta il numero di alunni, quindi ptratio diminuisce.
med_rooms_house = bost.groupby("medv")["rm"].mean().round(2)
print (med_rooms_house)
correlation2 = bost["medv"].corr(bost["rm"])
print (correlation2) #coeff. correlazione positivo = sono direttamente proporzionali, se aumenta il numero delle stanze,
#la casa costa di più
```

```python
#• Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? 
#E delle tasse?
correlation3 = bost["rad"].corr(bost["age"])
print (correlation3) #coeff. positivo = le prime case sono state costruite vicino all'autostrada
correlation4 = bost["rad"].corr(bost["rm"])
print (correlation4) #coeff. negativo = più sono vicine all'autostrada meno stanze ci sono
correlation5 = bost["rad"].corr(bost["tax"])
print (correlation5) #coeff. positivo (vicino a uno) le tasse crescono in proporzione all'avvicinarsi all'autostrada
```

```python
#Abbiamo un DataFrame di dipendenti: 
employees_df = pd.DataFrame({ 'employee_id': [101, 102, 103, 104, 105], 
                             'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
                             'department_id': [1, 2, 1, 2, 3] })
#E un DataFrame di dipartimenti: 
departments_df = pd.DataFrame({ 'department_id': [1, 2, 3], 
                               'department_name': ['HR', 'IT', 'Finance'], 
                               'location': ['New York', 'San Francisco', 'Chicago'] })
#• Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni 
#sia sui dipendenti che sui dipartimenti, usando la funzione .merge() 
#• Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente; 
#poi unire i due dataset mediante il metodo .join() • Ci sono differenze nel risultato? Quali? Perché?
```

```python
#• Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni 
#sia sui dipendenti che sui dipartimenti, usando la funzione .merge() 
pd.merge(employees_df,departments_df, on = "department_id", how = "inner")
```

```python
#• Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente
employees_df.set_index("department_id", inplace=True)
departments_df.set_index("department_id", inplace=True)
```

```python
display (employees_df)
display (departments_df)
```

```python
#poi unire i due dataset mediante il metodo .join() • Ci sono differenze nel risultato? Quali? Perché?
df_joined = employees_df.join(departments_df, how="inner")
display(df_joined)
#l'unica differenza è che in questo caso l'indice è quello che abbiamo settato (department_id) il resto delle colonne 
#e delle righe sono uguali.
```

```python
#Dal database AdventureWorksDW importiamo le tabelle dimemployee e dimemployeesalesterritory come DataFrame 
#• Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey 
#• Controlliamo la dimensione del DataFrame risultante: è quella attesa? 
#• Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante 
#della join precedente, usando le colonne SalesTerritoryKey 
#• Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region) 
#• Valutiamo la media del BaseRate per ogni paese: ci sono differenze?
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(override=True, dotenv_path="credenziali.env")

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(username, password, host, dbname)
db_engine = sqlalchemy.create_engine(connection_string)
query1 = "SELECT * FROM dimemployee"
dimemployee_df = pd.read_sql(query1, db_engine)
display(dimemployee_df)
query2 = "SELECT * FROM dimemployeesalesterritory"
dimemployeesalesterritory_df = pd.read_sql(query2, db_engine)
display(dimemployeesalesterritory_df)
```

```python
#• Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey
#• Controlliamo la dimensione del DataFrame risultante: è quella attesa?
employee_joined = pd.merge(dimemployee_df,dimemployeesalesterritory_df, on = "EmployeeKey", how = "left")
display(employee_joined)
```

```python
#• Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante 
#della join precedente, usando le colonne SalesTerritoryKey 
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(override=True, dotenv_path="credenziali.env")

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(username, password, host, dbname)
db_engine = sqlalchemy.create_engine(connection_string)
query3 = "SELECT * FROM dimsalesterritory"
dimsalesterritory_df = pd.read_sql(query3, db_engine)
display(dimsalesterritory_df)
```

```python
territory_joined = pd.merge(employee_joined,dimsalesterritory_df, on = "SalesTerritoryKey", how = "left")
display(territory_joined)
print(territory_joined.columns)
```

```python
#• Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)
dipendenti_per_country = territory_joined.groupby("SalesTerritoryCountry")["EmployeeKey"].count()
print (dipendenti_per_country)
```

```python
dipendenti_per_regione = territory_joined.groupby("SalesTerritoryRegion")["EmployeeKey"].count()
print (dipendenti_per_regione)
```

```python
#• Valutiamo la media del BaseRate per ogni paese: ci sono differenze?
media_baserate_per_country = territory_joined.groupby("SalesTerritoryCountry")["BaseRate"].mean()
print (media_baserate_per_country)
```
