```python
#Memorizza e stampa tutti i fattori di un numero dato in input (cioè tutti i numeri 
#primi che, moltiplicati tra di loro, danno il numero di partenza).
# Esempio:
# • input: 150
# • output: [2, 3, 5, 5]
num = int(input("Numero da scomporre in fattori primi"))
fattori = []
divisore = 2
while num > 1: #Continua a cercare fattori primi finché non abbiamo ridotto il numero a 1
    if num % divisore == 0:
        fattori.append(divisore)
        num = num // divisore
    else:
        divisore += 1
print(fattori)
```
